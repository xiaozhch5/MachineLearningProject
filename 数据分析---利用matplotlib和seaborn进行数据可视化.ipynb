# 准备工作

引入如下设置：

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')

large = 22; mid = 16; small = 12
params = {'axes.titlesize': large,
         'legend.fontsize': mid,
         'figure.figsize': (16, 10),
         'axes.labelsize': mid,
         'axes.titlesize': mid,
         'xtick.labelsize': mid,
         'ytick.labelsize': mid,
         'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
%matplotlib inline

# version
print(mpl.__version__)
print(sns.__version__)
```

# 关联

关联图标用于可视化两个或跟多变量之间的关系。

## 散点图

散点图是用于研究两个变量之间关系的经典的和基本你的图标。如果数据中有多个组，则可能需要以不同颜色可视化每个组。在matplotlib中，可使用<span class="girk">plt.scatterplot()</span>方便地执行。

```python
# 数据读取
midwest = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/midwest_filter.csv")
```

```python
# 数据处理
# 找出midwest中category下有哪几种不同的种类
categories = np.unique(midwest['category'])

# 对不同种类的category赋予一种颜色
colors = [plt.cm.tab10(i/float(len(categories) - 1)) for i in range(len(categories))]
plt.figure(figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')

for i, category in enumerate(categories):  # enumerate用法可以参考：https://love.lrting.top/archives/1003
    plt.scatter('area', 'poptotal',
               data=midwest.loc[midwest.category==category, :],
               s=20, cmap=colors[i], label=str(category))
    
plt.gca().set(xlim=(0.0, 0.1), ylim=(0, 90000),
              xlabel='Area', ylabel='Population')

plt.xticks(fontsize=12); plt.yticks(fontsize=12)
plt.title("Scatterplot of Midwest Area vs Population", fontsize=22)
plt.legend(fontsize=12)    
plt.show() 
```

## 相关图

相关图用于直观地查看给定数据框（或者二维数组）中所有可能的数值变量对支架的相关度量。

```python
# 导入数据
dataframe = pd.read_csv("https://github.com/selva86/datasets/raw/master/mtcars.csv")
```

```python
dataframe
```

```python
# 画图
plt.figure(figsize=(12, 10), dpi=80)
sns.heatmap(dataframe.corr(), xticklabels=dataframe.corr().columns, yticklabels=dataframe.corr().columns, cmap='RdYlGn', center=0, annot=True)
```

## 矩阵图

矩阵图是探索性分析中最爱，用于理解所有可能的数值变量对之间的关系。它是双变量分析的必备工具。

```python
# 导入数据
data_iris = sns.load_dataset('iris')

# 画图
plt.figure(figsize=(10, 8), dpi=80)
sns.pairplot(data_iris, kind="scatter", hue="species", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
```
